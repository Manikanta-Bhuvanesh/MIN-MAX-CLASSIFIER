```python
import pandas as pd 
import numpy as np
import statistics
from sklearn.model_selection import train_test_split
import random
from scipy import stats as s
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler

data=pd.read_csv('Crop_recommendation.csv')

class New:
  import pandas as pd 
  from scipy import stats as s
  X=[]
  Y=[]
  target = []
  numclass=[]
  def fi(self,X,Y):
    self.X=X
    self.Y=Y
    self.target=Y.unique()
    self.numclass=X.shape

  def pre(self,X2):
    n=self.X
    maxa={}
    mina={}
    normaldata = pd.concat([n, X2], axis=1, ignore_index=True)
    scale=MinMaxScaler(feature_range=(0, 1000))
    colnames=list(normaldata.columns)
    p=scale.fit_transform(normaldata.values)
    normaldata[colnames]=p
    n=normaldata.iloc[1:self.numclass[0],]
    X1=normaldata.iloc[self.numclass[0]:,]
    data = pd.concat([n, self.Y], axis=1, ignore_index=True)
    
    for i in self.target:
      temp=data[Y==i]
      temp=temp.iloc[: , :-1]
      maxa[i]=temp.max()
      mina[i]=temp.min()
      
    y_predict=[]
    for l in range(X1.shape[0]):
      predict=[]
      for j in self.target:
        for k in range(X1.shape[1]):
          if maxa[j][k]>=X1.iloc[l,k] and mina[j][k]<=X1.iloc[l,k]:
            predict.append(j)
      y_predict.append((s.mode(predict)[0]))  
    return y_predict
```

```python
X = data.drop("label", axis=1)
Y = data["label"]
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2)
model=New()
model.fi(X_train,Y_train)
y_predict=model.pre(X_test)
print(accuracy_score(Y_test,y_predict)*100)
```
