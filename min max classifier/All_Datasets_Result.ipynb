```python
import pandas as pd 
from scipy import stats as s
import pandas as pd 
import numpy as np
import statistics
from sklearn.model_selection import train_test_split
import random
from scipy import stats as s
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt
import time
full=[]
fullac=[]
```

```python
class New:
  data1 = []
  data=[]
  numclass=[]
  max={}
  min={}
  target = []
  def fi(self,X,Y):
    self.data = pd.concat([X, Y], axis=1, ignore_index=True)
    self.data1=X
    self.target=Y.unique()
    for i in self.target:
      temp=self.data[Y==i]
      temp=temp.iloc[: , :-1]
      self.max[i]=temp.max()
      self.min[i]=temp.min()

  def pre(self,X):
    y_predict=[]
    for l in range(X.shape[0]):
      predict=[]
      for j in self.target:
        for k in range(X.shape[1]):
          if self.max[j][k]>=X.iloc[l,k] and self.min[j][k]<=X.iloc[l,k]:
            predict.append(j)
      y_predict.append((s.mode(predict)[0]))  
    return y_predict
```

```python

from sklearn.neighbors import KNeighborsClassifier 
start_time = time.time()
knn=KNeighborsClassifier()
knn.fit(X_train,Y_train)
A=knn.predict(X_test)
Time.append(time.time() - start_time)
knn_accuracy=accuracy_score(Y_test,A)*100
Names.append('knn_algorithm')
Accuracy.append(knn_accuracy)

from sklearn.svm import SVC
start_time = time.time()
svm=SVC()
svm.fit(X_train,Y_train)
s=svm.predict(X_test)
Time.append(time.time() - start_time)
svm_accuracy=accuracy_score(Y_test,s)*100
Names.append('SVM_algorithm')
Accuracy.append(svm_accuracy)

from sklearn.naive_bayes import GaussianNB
start_time = time.time()
naive = GaussianNB()
naive.fit(X_train,Y_train)
n=naive.predict(X_test)
Time.append(time.time() - start_time)
naivebayes_accuracy=accuracy_score(Y_test,n)*100
Names.append('Naviebayes_algorithm')
Accuracy.append(naivebayes_accuracy)

from sklearn.tree import DecisionTreeClassifier
start_time = time.time()
tree=DecisionTreeClassifier()
tree.fit(X_train,Y_train)
c=tree.predict(X_test)
Time.append(time.time() - start_time)
DecisionTree_accuracy=accuracy_score(Y_test,c)*100
Names.append('DecisionTree_algorithm')
Accuracy.append(DecisionTree_accuracy)

from sklearn.ensemble import RandomForestClassifier
start_time = time.time()
random=RandomForestClassifier()
random.fit(X_train,Y_train)
r=random.predict(X_test)
Time.append(time.time() - start_time)
RandomForest_accuracy=accuracy_score(Y_test,r)*100
Names.append('RandomForest_algorithm')
Accuracy.append(RandomForest_accuracy)

full.append(Time)
fullac.append(Accuracy)
plt.figure(figsize=[8,5],dpi = 100)
plt.xlabel("Algorithms")
plt.ylabel("Accuracy")
plt.barh(Names,Accuracy)
for index, value in enumerate(Accuracy):
    plt.text(value, index,str(value))
plt.show()

plt.figure(figsize=[8,5],dpi = 100)
plt.xlabel("Algorithms")
plt.ylabel("Time")
plt.barh(Names,Time)
for index, value in enumerate(Time):
    plt.text(value, index,str(value))
plt.show()
```
