```python
import pandas as pd 
import numpy as np
from statistics import mode
from sklearn.model_selection import train_test_split
import random
from scipy import stats as s
from sklearn.preprocessing import MinMaxScaler
```

```python
data=pd.read_csv('Crop_recommendation.csv')
data1=data.drop("label", axis=1)
```

```python
numclass=data1.shape
```

```python
target=data.label.unique()
max_dic={}
min_dic={}
for i in target:
    temp=data[data.label==i]
    temp=temp.drop("label", axis=1)
    max_dic[i]=temp.max()
    min_dic[i]=temp.min()
print(max_dic)
print(min_dic)
```

```python
y_predict=[]
for l in range(numclass[0]):
    predict=[]
    for j in target:
        for k in range(numclass[1]):
            if max_dic[j][k]>=data1.iloc[l,k] and min_dic[j][k]<=data1.iloc[l,k]:
                predict.append(j)
    print(predict)
    print(mode(predict))
    y_predict.append((mode(predict)))  
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(data["label"],y_predict)*100)
```
