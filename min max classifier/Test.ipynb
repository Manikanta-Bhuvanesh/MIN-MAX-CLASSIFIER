```python
import pandas as pd 
from scipy import stats as s
import pandas as pd 
import numpy as np
import statistics
from sklearn.model_selection import train_test_split
import random
from scipy import stats as s
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score
```

```python
class New:
  data1 = []
  data=[]
  numclass=[]
  max={}
  min={}
  target = []
  def fi(self,X,Y):
    self.data = pd.concat([X, Y], axis=1, ignore_index=True)
    self.data1=X
    self.target=Y.unique()
    for i in self.target:
      temp=self.data[Y==i]
      temp=temp.iloc[: , :-1]
      self.max[i]=temp.max()
      self.min[i]=temp.min()

  def pre(self,X):
    y_predict=[]
    for l in range(X.shape[0]):
      predict=[]
      for j in self.target:
        for k in range(X.shape[1]):
          if self.max[j][k]>=X.iloc[l,k] and self.min[j][k]<=X.iloc[l,k]:
            predict.append(j)
      y_predict.append((s.mode(predict)[0]))  
    return y_predict
```

```python
data=pd.read_csv('diabetes2.csv')
data.describe()
```

```python
X = data.drop("Outcome", axis=1)
Y = data["Outcome"]
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, Y_train, Y_test=train_test_split(X,Y,test_size=0.2)
model=New()
model.fi(X_train,Y_train)
y_predict=model.pre(X_test)
print(accuracy_score(Y_test,y_predict)*100)
```
